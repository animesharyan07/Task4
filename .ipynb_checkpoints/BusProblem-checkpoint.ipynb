## Problem Statement
##### A school is preparing a trip for 400 students. The transportation company has 10 buses of 50 seats each and 8 buses of 40 seats but only has 9 drivers available. The rental cost for a large bus is 800 and 600 for a small bus. Calculate how many buses of each type should be used for the trip for the least possible cost.    Kindly refer to the steps for your understanding inside the directory[prblm.pdf]


```python
from pulp import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define the LP problem
prob = LpProblem("School_Trip_Min_Cost", LpMinimize)
```

```python
# Decision variables
x = LpVariable("Large_Buses", lowBound=0, upBound=10, cat=LpInteger)
y = LpVariable("Small_Buses", lowBound=0, upBound=8, cat=LpInteger)
```

```python
# Objective function: minimize cost
prob += 800 * x + 600 * y, "Total_Cost"
```

```python
# Constraints
prob += 50 * x + 40 * y >= 400, "Seats_Constraint"
prob += x + y <= 9, "Drivers_Constraint"
```

```python
# Solve the problem
prob.solve()
```

```python
# Extract solution
x_val = x.varValue
y_val = y.varValue
total_cost = value(prob.objective)

print(f"Large Buses Used: {x_val}")
print(f"Small Buses Used: {y_val}")
print(f"Total Cost: ₹{total_cost}")
```

```python
#lets Visualize the problem

x_vals = np.arange(0, 11, 1)
y_vals = np.arange(0, 9, 1)
X, Y = np.meshgrid(x_vals, y_vals)

```

```python
# Constraint regions
Z1 = 50 * X + 40 * Y   # Seats
Z2 = X + Y             # Drivers
```

```python
# Plot the feasible region
plt.figure(figsize=(10, 6))
```

```python
# Shade feasible area manually
feasible_x = []
feasible_y = []
for xi in x_vals:
    for yi in y_vals:
        if 50*xi + 40*yi >= 400 and xi + yi <= 9:
            feasible_x.append(xi)
            feasible_y.append(yi)
plt.scatter(feasible_x, feasible_y, color='lightgreen', label='Feasible Region')
```

```python
# Constraint lines
y1 = (400 - 50 * x_vals) / 40
y2 = 9 - x_vals
plt.plot(x_vals, y1, label='50x + 40y = 400', color='blue')
plt.plot(x_vals, y2, label='x + y = 9', color='red')
```

```python

# Solution point
plt.plot(x_val, y_val, 'ro', label='Optimal Solution')
plt.text(x_val+0.2, y_val, f'({int(x_val)}, {int(y_val)})', fontsize=12)
```

```python
# Bounds
plt.xlim(0, 10.5)
plt.ylim(0, 8.5)
```

```python


# Labels
plt.xlabel("Large Buses (x)")
plt.ylabel("Small Buses (y)")
plt.title("Feasible Region and Optimal Solution")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
